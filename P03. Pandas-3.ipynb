# Visualization

```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x=[-2,5,12]
x
```

```python
y=[-4,6,10]
y
```

```python
fig=plt.figure(figsize=(10,4))
plt.plot(x,y)
```

```python
fig=plt.figure(figsize=(15,5))
plt.plot(x,y)
plt.xlim(-5,15)
plt.ylim(-5,15)
plt.xlabel('Sale',size=15)
plt.ylabel('Profit',size=15)
plt.title('line plot',size =15)
plt.suptitle('Profit Figures', size=25, y=1.03)
```
